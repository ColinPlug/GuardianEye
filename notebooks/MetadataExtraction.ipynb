```python
install Pillow as PIL
from PIL import Image
from PIL.ExifTags import TAGS
import os

def vs(input_pad, output_pad):
    try:
        img = Image.open(input_pad)
        print(f"--- Analyse van: {input_pad} ---")

        # Metadata extraheren
        exif_data = img.getexif()
        if not exif_data:
            print("No metadata found!")
        else:
            for tag_id, value in exif_data.items():
                tag_name = TAGS.get(tag_id, tag_id)
                print(f"{tag_name:20}: {value}")

        # Strip sensitive tags (anonimise)
        data = list(img.getdata())
        clean_img = Image.new(img.mode, img.size)
        clean_img.putdata(data)

        # Save without metadata
        clean_img.save(output_pad)
        print(f"\n[Success] Sensitive data deleted. Saved as: {output_pad}")

    except Exception as e:
        print(f"Fout bij verwerken: {e}")
```
